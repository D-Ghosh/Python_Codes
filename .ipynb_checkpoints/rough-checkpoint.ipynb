```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# System parameters
omega_0 = 1.0  # TLS transition frequency
gamma_0 = 0.5  # Decay rate (Markovian approximation)
tlist = np.linspace(0, 50, 500)  # Time range

# Operators
sigma_z = sigmaz()
sigma_p = sigmap()
sigma_m = sigmam()

# Hamiltonian (TLS alone)
H = (omega_0 / 2) * sigma_z

# Initial state (excited state)
psi0 = basis(2, 1)
rho0 = psi0 * psi0.dag()

# Function to compute the tilted Liouvillian
def tilted_master_eq(chi):
    """
    Construct the Liouvillian with counting field χ
    """
    L0 = liouvillian(H, [np.sqrt(gamma_0) * sigma_m])
    
    # Modify the emission term with counting field e^(iχ)
    J = gamma_0 * (np.exp(1j * chi) * sprepost(sigma_m, sigma_p) - 0.5 * spre(sigma_p * sigma_m) - 0.5 * spost(sigma_p * sigma_m))
    
    return L0 + J

# Solve the master equation for different χ values
chi_values = np.linspace(-1, 1, 50)  # Range of χ values
Z_chi_t = []

for chi in chi_values:
    L_chi = tilted_master_eq(chi)
    result = mesolve(L_chi, rho0, tlist, [], [])
    Z_chi_t.append([np.real(np.trace(result.states[i].full())) for i in range(len(tlist))])


Z_chi_t = np.array(Z_chi_t)

# Compute cumulants
F_chi = np.log(Z_chi_t) / tlist[-1]  # Log moment generating function (scaled)
current = np.gradient(F_chi, chi_values, axis=0)  # First derivative
noise = np.gradient(current, chi_values, axis=0)  # Second derivative

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(chi_values, current[:, -1], label="Current (Mean Emission Rate)")
plt.plot(chi_values, noise[:, -1], label="Noise (Fluctuations)")
plt.axvline(0, color="black", linestyle="--", linewidth=0.8)
plt.xlabel(r"Counting Field $\chi$")
plt.ylabel("Cumulants")
plt.legend()
plt.title("Full Counting Statistics of Emissions")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# System parameters
omega_0 = 1.0  # TLS transition frequency
gamma_0 = 1.0  # Decay rate
tlist = np.linspace(0, 50, 500)  # Time range

# Operators
sigma_z = sigmaz()
sigma_p = sigmap()
sigma_m = sigmam()

# Hamiltonian (TLS alone)
H = (omega_0 / 2) * sigma_z

# Initial state (excited state)
psi0 = basis(2, 1)
rho0 = psi0 * psi0.dag()

# Function to compute the tilted Liouvillian
def tilted_master_eq(chi):
    """
    Construct the Liouvillian with counting field χ
    """
    L0 = liouvillian(H, [np.sqrt(gamma_0) * sigma_m])
    
    # Modify the emission term with counting field e^(iχ)
    J = gamma_0 * (sprepost(sigma_m, sigma_p) * np.exp(1j * chi) - 0.5 * spre(sigma_p * sigma_m) - 0.5 * spost(sigma_p * sigma_m))
    
    return L0 + J

# Solve the master equation for different χ values
chi_values = np.linspace(-np.pi, np.pi, 50)  # Fourier sampling range
Z_chi_t = []

for chi in chi_values:
    L_chi = tilted_master_eq(chi)
    result = mesolve(L_chi, rho0, tlist, [], [])
    Z_chi_t.append([expect(sigma_p * sigma_m, result.states[i]) for i in range(len(tlist))])  # Corrected expectation

Z_chi_t = np.array(Z_chi_t)

# Compute probability distribution function using inverse Fourier transform
n_values = np.arange(0, 20)  # Number of emitted photons
P_n_t = []

for t_idx in range(len(tlist)):
    P_n = []
    for n in n_values:
        integral = np.trapz(np.exp(-1j * n * chi_values) * Z_chi_t[:, t_idx], chi_values)
        P_n.append(np.real(integral / (2 * np.pi)))  # Normalize
    P_n_t.append(P_n)

P_n_t = np.array(P_n_t)

# Plot probability distribution for a fixed time
time_index = -1  # Choose final time step for visualization
plt.figure(figsize=(8, 5))
plt.bar(n_values, P_n_t[time_index], color='b', alpha=0.7, label=f"t = {tlist[time_index]:.1f}")
plt.xlabel("Number of Emissions (n)")
plt.ylabel("Probability P(n, t)")
plt.title("Probability Distribution of Emission Events")
plt.legend()
plt.grid()
plt.show()
```
