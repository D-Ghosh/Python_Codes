```python
from qutip import *
import numpy as np

# Basic states
g = basis(2, 0)
e = basis(2, 1)

# Tensor basis
ge = tensor(g, e)
eg = tensor(e, g)
gg = tensor(g, g)
ee = tensor(e, e)

# Symmetric and antisymmetric states
s = (eg + ge).unit()
a = (eg - ge).unit()

# Projectors
P_ee = ket2dm(ee)
P_gg = ket2dm(gg)
P_ss = ket2dm(s)
P_aa = ket2dm(a)

# Define full Hilbert space size
dim = P_ee.shape[0]

# Hamiltonian
ħ = 1.0
x0 = 1.0
X12 = 0.5
Has = ħ * x0 * (P_ee - P_gg) + ħ * X12 * (P_ss - P_aa)

# Decay parameters
C = 1.0
C12 = 0.8

# Decay jump operators (need to act on full Hilbert space)
# |g⟩⟨s| and |s⟩⟨e|
L_gs = tensor(g, g) * s.dag()
L_se = s * tensor(e, e).dag()
L_s = np.sqrt(C + C12) * (L_gs + L_se)

L_ga = tensor(g, g) * a.dag()
L_ae = a * tensor(e, e).dag()
L_a = np.sqrt(C - C12) * (L_ga - L_ae)

# Initial state: both atoms excited
rho0 = ket2dm(ee)

# Time list
tlist = np.linspace(0, 10, 200)

# Solve
result = mesolve(Has, rho0, tlist, [L_s, L_a], [])

# Plot expectation of some observable (e.g., excitation population)
expect_ee = expect(P_ee, result.states)

import matplotlib.pyplot as plt
plt.plot(tlist, expect_ee, label='P_ee')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt

# Basic states
g = basis(2, 0)
e = basis(2, 1)

# Tensor basis
ge = tensor(g, e)
eg = tensor(e, g)
gg = tensor(g, g)
ee = tensor(e, e)

# Symmetric and antisymmetric states
s = (eg + ge).unit()
a = (eg - ge).unit()

# Projectors
P_ee = ket2dm(ee)
P_gg = ket2dm(gg)
P_ss = ket2dm(s)
P_aa = ket2dm(a)

# Hamiltonian
ħ = 1.0
x0 = 1.0
X12 = 0.5
Has = ħ * x0 * (P_ee - P_gg) + ħ * X12 * (P_ss - P_aa)

# Decay parameters
C = 1.0
C12 = 0.8

# Collapse operators
L_gs = tensor(g, g) * s.dag()
L_se = s * tensor(e, e).dag()
L_s = np.sqrt(C + C12) * (L_gs + L_se)

L_ga = tensor(g, g) * a.dag()
L_ae = a * tensor(e, e).dag()
L_a = np.sqrt(C - C12) * (L_ga - L_ae)

# Initial state
rho0 = ket2dm(ee)
tlist = np.linspace(0, 10, 200)

# Solve master equation
result = mesolve(Has, rho0, tlist, [L_s, L_a], [])

# Spin current operator (i[σ1⁺σ2⁻ - σ1⁻σ2⁺])
sigmap = sigmap()
sigmam = sigmam()
J_spin_op = 1j * (tensor(sigmap, sigmam) - tensor(sigmam, sigmap))
spin_current = expect(J_spin_op, result.states)

# Heat current: J_heat = Tr[H D[L](ρ)]
def lindblad_heat_current(H, rho, L):
    D = L * rho * L.dag() - 0.5 * (L.dag() * L * rho + rho * L.dag() * L)
    return np.real((H * D).tr())

heat_current = [lindblad_heat_current(Has, state, L_s) +
                lindblad_heat_current(Has, state, L_a)
                for state in result.states]

# Plot all
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(tlist, spin_current, label='Spin Current')
plt.xlabel('Time')
plt.ylabel('⟨J_spin⟩')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(tlist, heat_current, label='Heat Current', color='red')
plt.xlabel('Time')
plt.ylabel('J_heat')
plt.legend()

plt.tight_layout()
plt.show()
```
